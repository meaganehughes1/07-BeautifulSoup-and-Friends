```python
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get("https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States")
doc = BeautifulSoup(response.text)
```

```python
doc.find('b').get_text()
```

```python
presidents = doc.select('b')
rows = []

for president in presidents:
    row = {}
    # Get the text inside of b, if no text say none
    row['president name'] = president.get_text('b', None)
    rows.append(row)
len(rows)

#taking a slice of this list for the number of presidents, because "None" doesn't seem to be working to do this - must be other 'b' elements on the page?
presrows = rows[0:47]
```

```python
import pandas as pd
df = pd.json_normalize(presrows)
df
```

```python
df.to_csv("presidents.csv", index=False)
```
